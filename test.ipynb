```python
from graph import Node
import networkx as nx

G = nx.Graph()

# Create nodes using the Node class and add them to the graph
node1 = Node(1, 'red', False, [(1, 2), (1, 3)])
node2 = Node(2, 'blue', True, [(2, 3)])

nodes_to_add = [node1, node2]

for node in nodes_to_add:
    node_id, node_attrs = node.node_view()
    G.add_node(node_id, **node_attrs)
    G.add_edges_from(node.edges)

# Print the nodes and edges of the graph
print("Nodes:", G.nodes(data=True))
print("Edges:", G.edges())
```

```python
from graph_utils import convert_maze

maze_file_path = 'test1.csv'
maze_graph = convert_maze(maze_file_path)
print("Nodes:", maze_graph.nodes())
print("Edges:", maze_graph.edges())
```

```python
import matplotlib.pyplot as plt

pos = {node_id: (int(node_id.split(',')[0]), int(node_id.split(',')[1])) for node_id in maze_graph.nodes()}

# Draw the maze graph using NetworkX and matplotlib with grid layout
nx.draw(maze_graph, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=8)
plt.title('Maze Graph (Grid Layout)')
plt.show()
```
