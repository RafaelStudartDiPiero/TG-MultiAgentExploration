```python
import networkx as nx

def rename_nodes(graph):
    # Criar um mapeamento para renomear os nós
    mapping = {node: node.replace(',', '_') for node in graph.nodes()}
    # Renomear os nós no gráfico
    nx.relabel_nodes(graph, mapping, copy=False)

# Criar a árvore a partir do gráfico
tree = custom_dfs_tree(graph, "0,0")

# Renomear os nós
rename_nodes(tree)

# Imprimir nós e arestas da árvore renomeada
print(tree.nodes)
print(tree.edges)

# Exibir o gráfico
display_graph(tree, "0_0")
```
