```python

dfs_tree = analyze_graph(maze_graph,"6,6")

pos = nx.nx_agraph.graphviz_layout(dfs_tree, prog="dot", root="6,6")

# Draw the DFS tree with custom styling
plt.figure(figsize=(10, 8))
nx.draw(dfs_tree, pos, with_labels=True, node_size=500, node_color="lightblue", 
        font_size=10, font_weight="bold", edge_color="gray")

# Display the graph
plt.title("DFS Tree Layout")
plt.show()
```
